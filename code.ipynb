```python
wines.dropna(subset = ['Harvest date','DOY'], inplace = True)
wines.reset_index(drop=True, inplace = True)
```

```python
def jour(nb_jour):
    colonnes = {}
    #creer dataframe vide avec nombre de jour qu'on veut avant harvest date en colonne
    for i in range(nb_jour):
        colonnes[f"colonne_{i+1}"] = []
    df = pd.DataFrame(colonnes)
    for index in range(len(wines)): 
        meteo = weather[wines.iat[index,5]]
        harvest_date = wines.iat[index,6]
        year = int(harvest_date.year) #recup l'année
        jour_exact = int(wines.iat[index,7])
        
        #ajouter pour chaque jour le la temparature moyenne donc a recup le harvest date et 
        analyse = jour_exact - nb_jour #la ou on commence 
        result = meteo.loc[(meteo['Year'] == year) & (meteo['DOY'].isin(range(analyse,jour_exact)))]['Tave']
        liste_temp =list(result)
        if len(liste_temp) != nb_jour :
            liste_temp = [float('nan')]*nb_jour
        df.loc[index] = liste_temp
    return df
```
